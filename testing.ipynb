```python
from uuid import UUID

import core
```

```python
def print_type(obj) -> None:
    print(type(obj))  # noqa: T201
```

# Operational Schema


```python
# Get database session (operational is default)
db = core.dependencies.get_db_session()
```

```python
# Query a single project
project_mi = core.crud.operational.projects.get_project(
    db=db, project_id=UUID("623ddc81-ed4c-4e56-b9ca-a4a9be238a8b")
)
print(project_mi)  # noqa: T201
print_type(project_mi)
```

```python
# Get SQL string
project_sql = project_mi.sql_string()
print(project_sql)  # noqa: T201
print_type(project_sql)
```

```python
# Get dictionary
project_dict = project_mi.dictionary()
print(project_dict)  # noqa: T201
print_type(project_dict)
```

```python
# Get SQLAlchemy model
project_model = project_mi.model()
print(project_model)  # noqa: T201
print_type(project_model)
```

```python
# Get Pandas Series
project_s = project_mi.pandas_series()
print(project_s.head(5))  # noqa: T201
print_type(project_s)
```

```python
# Get Pandas DataFrame
project_df = project_mi.pandas_dataframe()
_ = display(project_df)
print_type(project_df)
```

```python
# Get Polars DataFrame (expecting warning)
project_df = project_mi.polars_dataframe()
_ = display(project_df)
print_type(project_df)
```

```python
# Re-run crud function preparing for Polars
project_mi_lazy = core.crud.operational.projects.get_project(
    db=db, project_id=UUID("623ddc81-ed4c-4e56-b9ca-a4a9be238a8b"), return_query=True
)
print(project_mi_lazy)  # noqa: T201
print_type(project_mi_lazy)

# Get Polars DataFrame
project_df = project_mi_lazy.polars_dataframe()
_ = display(project_df)
print_type(project_df)
```

```python
# Expect error when accessing items before query
try:
    _ = project_mi_lazy.model()
except Exception as e:
    print(e.__class__)  # noqa: T201
    print(e)  # noqa: T201
```

```python
# Call query_item() and repeat
project_mi.query_item()
print(project_mi.item)  # noqa: T201
print_type(project_mi.item)
```

```python
# Query projects
projects_ml = core.crud.operational.projects.get_projects(db=db)
print(projects_ml)  # noqa: T201
print_type(projects_ml)
```

```python
# Get SQL string
projects_sql = projects_ml.sql_string()
print(projects_sql)  # noqa: T201
print_type(projects_sql)
```

```python
# Get SQLAlchemy models
projects_models = projects_ml.models()[:3]
print(projects_models)  # noqa: T201
print_type(projects_models)
```

```python
# Get Pandas DataFrame
projects_df = projects_ml.pandas_dataframe().head(3)
_ = display(projects_df)
print_type(projects_df)
```

```python
# Get Polars DataFrame (expecting warning)
projects_df = projects_ml.polars_dataframe().head(3)
_ = display(projects_df)
print_type(projects_df)
```

```python
# Re-run crud function preparing for Polars
projects_ml_lazy = core.crud.operational.projects.get_projects(db=db, return_query=True)
print(projects_ml_lazy)  # noqa: T201
print_type(projects_ml_lazy)

# Get Polars DataFrame
project_df = projects_ml_lazy.polars_dataframe().head(3)
_ = display(project_df)
print_type(project_df)
```

```python
# Expect error when accessing items before query
try:
    _ = projects_ml_lazy.models()
except Exception as e:
    print(e.__class__)  # noqa: T201
    print(e)  # noqa: T201
```

```python
# Call query_items() and repeat
projects_ml_lazy.query_items()
print(projects_ml_lazy.models())  # noqa: T201
print_type(projects_ml_lazy.models())
```
